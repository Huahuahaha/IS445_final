```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('netflix_titles.csv', encoding='ISO-8859-1')

# 1. Content Type Distribution
type_counts = data['type'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Distribution of Content Types on Netflix')
plt.axis('equal')
plt.show()

# 2. Top 5 Countries by Netflix Content
top_countries = data['country'].value_counts().head(5)
plt.figure(figsize=(10, 5))
top_countries.plot(kind='bar', color='#ffcc99')
plt.title('Top 5 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()

# 3. Release Years of Netflix Content
recent_data = data[data['release_year'] >= 2010]  # Focus on the last decade
plt.figure(figsize=(10, 6))
plt.hist(recent_data['release_year'], bins=range(2010, 2023), color='#99ccff', alpha=0.7)
plt.title('Release Years of Netflix Content (Last Decade)')
plt.xlabel('Year')
plt.ylabel('Number of Titles Released')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
